```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Load data
obs = np.loadtxt('obs.csv', delimiter=',')
act = np.loadtxt('act.csv', delimiter=',')
```

```python
_ ,S, Vh = np.linalg.svd(obs, full_matrices=True)# [1:2]
print("eigenvalues: ", S)
Vh1 = Vh[[0, 1, 3, 4],:]
obs_low_dim = obs.dot(Vh1.T)
```

```python
plt.figure(figsize=(10, 5), dpi=200)
plt.plot(obs_low_dim)
plt.show()
```

```python
# 2. 分析周期性
start1 = 165
end1 = 240
plt.figure(figsize=(10, 5), dpi=300)
plt.plot(obs_low_dim[:,0])
plt.plot([start1,start1], [-10,12], 'r--')
plt.plot([end1,end1], [-10,12], 'r--')
start2 = 318
end2 = 390
plt.plot([start2,start2], [-10,12], 'g--')
plt.plot([end2,end2], [-10,12], 'g--')
plt.show()
```

```python
min_i = start1-10
min_j = end1+10
abs_min = np.inf

for i in range(start1 - 20, start1 + 20):
    for j in range(end1-10, end1+10):
        abs_val = np.sum(np.abs(obs_low_dim[i] - obs_low_dim[j]))
        if abs_val <= abs_min:
            min_i = i
            min_j = j
            abs_min = abs_val
print("min_i: ", min_i)
print("min_j: ", min_j)
print("abs_min: ", abs_val)
```
